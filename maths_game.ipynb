```
import time as t
import pandas as pd
import random as rnd
```

```
def main():

    # def fun.. which checks the usage MODE
    def check_mode():

        print('Enter Values input Mode')
        d = '{digits --> \'d\'}'
        t = '{tables --> \'t\'}'
        s = '{squares --> \'s\'}'
        mul_9 = '{squares --> \'mul_9\'}'
        mode = input(f'{d}\n{t}\  n{s}\n{mul_9}\n')

        if mode not in ('d','s','t','mul_9'):
            return check_mode()
        else:
            return mode

    # def fun.. which generates list contains range of values
    def val_list(mode):

        # con.. block runs when user choice --> 'd'
        if mode.lower() == 'd':

            row_dig = int(input('Enter no.of Digits for 1st Factor: '))
            col_dig = int(input('Enter no.of Digits for 2nd Factor: '))
            print('\n')

            row_limit = list(range(int('9' * row_dig)+1))
            col_limit = list(range(int('9' * col_dig)+1))

        # con.. block runs when user choice --> 't'
        elif mode.lower() == 't':

            row_limit = [int(i) for i in input('Enter Table name: ').split()]
            col_factorial = int(input('Enter last Column factorial: '))
            col_limit = list(range(col_factorial))

        # con.. block runs when user choice --> 's'
        elif mode.lower() == 's':

            square_lower_limit = int(input('Enter Sqr_lower_limit: '))
            square_upper_limit = int(input('Enter Sqr_upper_limit: '))
            row_limit = list(range(square_lower_limit, square_upper_limit+1))
            col_limit = row_limit

        # con.. block runs when user choice --> 'mul_9'
        elif mode.lower() == 'mul_9':
            num_dig_rng = int(input('Enter upper limit of no.of digits: '))

            row_limit = list(int(i * '9') for i in range(1,num_dig_rng+1))
            col_limit = list(range(int('9' * num_dig_rng)+1))


        return row_limit, col_limit, mode

    # def fun.. which picks 2 random factors for multiple
    def pick_rnd_fact(row_limit, col_limit):
        num = rnd.choice(row_limit)
        mul = rnd.choice(col_limit)
        if num < mul:
            num, mul = mul,num

        return num, mul

    # def fun.. which picks 1 random factors for squares
    def pick_rnd_square(row_limit):
        num = rnd.choice(row_limit)
        mul = num

        return num, mul

    # def fun.. which return actual value as result, user given response, time taken
    def question(num, mul):
        start_time = t.time()
        rsp = input(f'{num} * {mul} : ').lower()
        end_time = t.time()
        total_time = round(end_time - start_time, 2)
        result = num * mul

        if rsp == 'stop':
            return rsp, total_time, result
        elif rsp == 'skip':
            return rsp, total_time, result
        elif not(rsp.isdigit()):
            return question(num, mul)
        else:
            return int(rsp), total_time, result

    # def fun.. which compares the user input values and actual values
    def check_result(rsp, result):

        if rsp == result:
            print('CORRECT ANSWER !!\n')
        else:
            print(f'WRONG ANSWER !! \nCorrect Answer = {result}\n')

    # def fun.. which contains all above user def fun.. to run calculations
    def run():

        mode = check_mode()
        row_limit, col_limit, mode = val_list(mode)

        while True:
            if mode in ('d', 't', 'mul_9'):
                num, mul = pick_rnd_fact(row_limit, col_limit)
            elif mode in ('s'):
                num, mul = pick_rnd_square(row_limit)
            rsp, total_time, result = question(num, mul)
            if rsp == 'stop':
                break
            elif rsp == 'skip':
                continue
            check_result(rsp, result)

    run() # calling fun..








main() # calling fun..
```
